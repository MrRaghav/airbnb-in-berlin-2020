```python
#converting the text into lowercase
summaryReviews['comments'] = summaryReviews['comments'].str.lower()
```

```python
#blank values in the comments are being removed
summaryReviews = summaryReviews[summaryReviews['comments'].notnull()]
```

```python
# replacement of the new line character
summaryReviews['comments'] = summaryReviews['comments'].str.replace('\r\n', "")
```

```python
#numbers are being removed
summaryReviews['comments'] = summaryReviews['comments'].str.replace('\d+', '') 
```

```python
#importing necessary libraries for text data

import re
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
```

```python
#https://stackoverflow.com/questions/44751922/unable-to-remove-english-stopwords-from-a-dataframe
```

```python
#Although comments are in german too, but we focus on english
englishStopwords = stopwords.words("english")
summaryReviews['comments'] = summaryReviews['comments'].apply(lambda x: " ".join([i for i in x.split() 
                                                      if i not in (englishStopwords)]))
```

```python
# fixing multiple spaces thing
summaryReviews['comments'] = summaryReviews['comments'].str.replace('\s+', ' ')
```

```python
# fixing punctuation thing
summaryReviews['comments'] = summaryReviews['comments'].str.replace('[^\w\s]'," ")
```

```python
namesOfHost = summaryReviews[['host_name']].drop_duplicates()
namesOfHost = namesOfHost.host_name.str.lower().tolist()

len(namesOfHost)
```

```python
listedComments = summaryReviews.comments.tolist()

vec = CountVectorizer().fit(listedComments)
BOW = vec.transform(listedComments) #bag of words approach
addedWords = BOW.sum(axis=0)
frequencyOfWords = [(word, addedWords[0, idx]) for word, idx in vec.vocabulary_.items()]

dfOf_CVEC = pd.DataFrame.from_records(frequencyOfWords, columns= ['words', 'counts']).sort_values(by="counts", ascending=False)
dfOf_CVEC.head(3)
```

```python
import matplotlib.pyplot as plt
```

```python
#https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.imshow.html
#https://stackoverflow.com/questions/12473511/what-does-matplotlib-imshowinterpolation-nearest-do/12473913
#https://matplotlib.org/examples/images_contours_and_fields/interpolation_none_vs_nearest.html
#https://matplotlib.org/3.2.1/gallery/images_contours_and_fields/interpolation_methods.html

dictionaryOf_CVEC = dict(zip(dfOf_CVEC.words, dfOf_CVEC.counts))

wordcloud = WordCloud(width=671, height=353)
wordcloud.generate_from_frequencies(frequencies=dictionaryOf_CVEC)
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation="sinc")
plt.axis("off")
plt.show()
```
